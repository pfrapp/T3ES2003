## Aufgabe 2.2: Sprungantwort

Uebertragungsfunktion des Systems ist
$$G = \frac{1}{s+1}$$

```python
s = ctrl.TransferFunction.s
G = 1/(s+1)
print(G)

t, h = ctrl.step_response(G)

plt.figure('Sprungantwort')
plt.clf()
plt.plot(t, h)
plt.grid(True)
plt.xlabel('t')
plt.ylabel('h')
plt.show()
```

## Aufgabe 2.3: Impulsantwort

System
$$G=\frac{1}{s+1}$$


```python
G = ctrl.tf([1], [1, 1])

t, g = ctrl.impulse_response(G)

plt.figure('Impulsantwort')
plt.clf()
plt.plot(t, g)
plt.grid(True)
plt.xlabel('t')
plt.ylabel('g')
plt.show()
```

## Aufgabe 2.4: Blockdiagramme


```python
G1 = 2
G2 = ctrl.tf([4], [1, 0])
G3 = 1/ctrl.TransferFunction.s

G_parallel = G1 + G2
G_reihe = G3 * G_parallel

G_feedback = ctrl.tf([1], [1])
G = ctrl.feedback(G_reihe, G_feedback, -1)

print(G)
```
